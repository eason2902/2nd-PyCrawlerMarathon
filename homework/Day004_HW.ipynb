```python
import requests
from pprint import pprint
r = requests.get('http://ins-info.ib.gov.tw/opendata/json-05010111.aspx')
pprint(r.text)
```

```python
import json
j_doc = json.loads(r.text)
type(j_doc)
pprint(j_doc)
```

```python
pprint(j_doc[0])
pprint(j_doc[0]['INSURER_Name'])
print(j_doc[0].items())
for key,value in j_doc[0].items():
    print(key,value)
```

```python
import csv
column = []
for key in j_doc[0].keys():
    column.append(key)

data =[]
data.append(column)
for co in range(len(j_doc)):
    row= []
    for value in j_doc[co].values():
        row.append(value)
    data.append(row)
del data[-2:]
#print(data)
with open('../Data/ex4.csv','w',encoding='utf-8-sig',newline='') as csvfile:
    w =csv.writer(csvfile)
    w.writerows(data)
```
