### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text) #"html5lib"
row_data =[] 
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_='tag').text , d.find_all('a')[-1].text)
    row_data.append([d.find(class_="date").text,d.find(class_='tag').text , d.find_all('a')[-1].text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
from pprint import pprint
kinds =[] 
for rowtxt in row_data:
    if rowtxt[1] not in kinds:
        kinds.append(rowtxt[1])
row_dict ={}
for kind in kinds:
    titles = []
    for rowtxt in row_data:
        if rowtxt[1] == kind:
            titles.append(rowtxt[2])
        row_dict[ kind ] = titles
pprint(row_dict)
```

### ③ 哪一個類別的文章最多

```python
len_kind = {}
for key,value in row_dict.items():  #key指派為tuple[0]
        len_kind[key] = len(value)

max_num = 0
for key,value in len_kind.items():
    if value > max_num:
        max_num = value
        max_kind = key
print(max_kind)
```
