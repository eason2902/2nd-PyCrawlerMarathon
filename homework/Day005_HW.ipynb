```python
請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
```

```python
import requests
from pprint import pprint
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
pprint(response[:1000])
```

```python
import json
data= json.loads(response)
pprint(data)
```

```python
#1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
print('有{}筆資料'.format(len(data)))
print('有以下欄位:')
for key in data[0].keys():
    print(key,end=' ,')
```

```python
#2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for i in range(len(data)):
    print(data[i]['title'],data[i]['createdAt'],data[i]['commentCount'],data[i]['likeCount'])
```

```python
#3.計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
data_f = json.loads(response)

commentCount_f_sum = 0
likeCount_f_sum = 0
for i in range(len(data_f)):
    commentCount_f_sum += data_f[i]['commentCount']
    likeCount_f_sum += data_f[i]['likeCount']
commentCount_f_avg = commentCount_f_sum / len(data_f)
likeCount_f_avg = likeCount_f_sum / len(data_f)

commentCount_sum = 0
likeCount_sum = 0
for i in range(len(data)):
    commentCount_sum += data[i]['commentCount']
    likeCount_sum += data[i]['likeCount']
commentCount_avg = commentCount_sum / len(data)
likeCount_avg = likeCount_sum / len(data)
print('非熱門文章的平均留言人數為{}，平均按讚人數為{}'.format(commentCount_f_avg,likeCount_f_avg))
print('熱門文章的平均留言人數為{}，平均按讚人數為{}'.format(commentCount_avg,likeCount_avg))
```
